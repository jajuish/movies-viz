### Processing the movies dataset

```python
movies = pd.read_csv("./movies-1990-to-2017/Movie_Movies.csv")
movies = movies.iloc[:,1:]
#movies.tail()
```

##### Dividing the accolades into separate columns

```python
for i in range(movies.shape[0]):
    m = movies['Awards'][i]
    if type(m) != float: # this is to filter out the nan's
        print(i,m)
```

```python
n = movies.shape[0]
x = [0 for i in range(n)]
other_wins = x
other_nominations = x
Oscar_wins = x
Oscar_nominations = x
BAFTA_wins = x
BAFTA_nominations = x
Emmy_wins = x
Emmy_nominations = x
GG_wins = x
GG_nominations = x
```

```python
movies['OscarWins'] = Oscar_wins
movies['OscarNominations'] = Oscar_nominations
movies['BaftaWins'] = BAFTA_wins
movies['BaftaNominations'] = BAFTA_nominations
movies['EmmyWins'] = Emmy_wins
movies['EmmyNominations'] = Emmy_nominations
movies['GoldenGlobeWins'] = GG_wins
movies['GoldenGlobeNominations'] = GG_nominations
movies['OtherWins'] = other_wins
movies['OtherNominations'] = other_nominations
```

```python
for i in range(movies.shape[0]):
    m = movies['Awards'][i]
    if (type(m) != float): # this is to filter out the nan's
        if 'Nominated for ' in m:
            osc = re.match('Nominated for . Oscar',m)
            baf = re.match('Nominated for . BAFTA Film Award',m)
            em = re.match('Nominated for . Primetime Emmy',m)
            gg = re.match('Nominated for . Golden Globe',m)
            
            # if it is a 2-digit number
            if m[15] != ' ':
                nom = int(m[14]+m[15])
            else:
                nom = int(m[14])
            
            if osc != None:
                movies['OscarNominations'][i] = nom
            elif baf != None:
                movies['BaftaNominations'][i] = nom
                
        if 'nomination' in m:
            ind = m.find('nomination')
            
            #if the number of nominations is a 2-digit number
            if(m[ind-3]) != ' ' and (m[ind-3]) != '.' :
                movies['OtherNominations'][i] = int(m[ind-3]+m[ind-2])
            else:
                movies['OtherNominations'][i] = int(m[ind-2])
                
        if 'Won ' in m:
            osc = re.match('Won . Oscar',m)
            baf = re.match('Won . BAFTA Film Award',m)
            em = re.match('Won . Primetime Emmy',m)
            gg = re.match('Won . Golden Globe',m)
            
            #if it is a 2-digit number
            if m[5] != ' ':
                won = int(m[4]+m[5])
            else:
                won = int(m[4])
            
            if osc != None:
                movies['OscarWins'][i] = won
            elif baf != None:
                movies['BaftaWins'][i] = won
            elif em != None:
                movies['EmmyWins'][i] = won
            elif gg != None:
                movies['GoldenGlobeWins'][i] = won
                
        if 'win' in m:
            ind = m.find('win')
            
            # if the number of other wins is a 2-digit number
            if(m[ind-3]) != ' ' and (m[ind-3]) != '.':
                movies['OtherWins'][i] = int(m[ind-3]+m[ind-2])
            else:
                movies['OtherWins'][i] = int(m[ind-2])
```

```python
for i in range(movies.shape[0]):
    m = movies['Awards'][i]
    if (type(m) != float): # this is to filter out the nan's
        if 'Nominated for ' in m:
            osc = re.match('Nominated for . Oscar',m)
            baf = re.match('Nominated for . BAFTA Film Award',m)
            em = re.match('Nominated for . Primetime Emmy',m)
            gg = re.match('Nominated for . Golden Globe',m)
            
            # if it is a 2-digit number
            if m[15] != ' ':
                nom = int(m[14]+m[15])
            else:
                nom = int(m[14])
            
            if gg != None:
                movies['GoldenGlobeNominations'][i] = nom
            if em != None:
                movies['EmmyNominations'][i] = nom
```

###### Scaling and normalizing the feature 'Box Office'

```python
n = movies.shape[0]
empty = [0 for i in range(n)]
movies['BoxOfficeRevenue'] = empty
```

```python
for i in range(movies.shape[0]):
    m = movies['BoxOffice'][i]
    if type(m) != float: # this is to filter out the nan's
        x = m[1:]
        if x[-1] == 'k':
            x = x[:-1]
            movies['BoxOfficeRevenue'][i] = float(x)*1000
        else:
            y = ''
            for j in x:
                if j != ',':
                    y+=j
            movies['BoxOfficeRevenue'][i] = float(y)
```

###### Scaling/Normalizing the ratings

```python
movies['Rated'].unique()
```

```python
n = movies.shape[0]
x = ['' for i in range(n)]
movies['RatedNormalized'] = x
```

```python
for i in range(movies.shape[0]):
    m = movies['Rated'][i]
    if type(m) != float: # this is to filter out the nan's
        if m == 'Approved' or m == 'APPROVED' or m == 'Passed' or m == 'PASSED':
            movies['RatedNormalized'][i] = 'APPROVED'
        elif m == 'NOT RATED' or m == 'UNRATED' or m == 'Not Rated' or m == 'Unrated' or m == 'NR' or m == 'Not rated':
            movies['RatedNormalized'][i] = 'NR'
        elif m == 'GP':
            movies['RatedNormalized'][i] = 'PG'
        else:
            movies['RatedNormalized'][i] = m
```

```python
movies['RatedNormalized'].unique()
```

###### scale/normalize the runtime

```python
n = movies.shape[0]
x = [0 for i in range(n)]
movies['RuntimeInMinutes'] = x
```

```python
for i in range(n):
    m = movies['Runtime'][i]
    if type(m) != float:
        time = 0
        if 'h' in m:
            h2 = re.match('[0-9][0-9] h',m) # 2 digit
            h1 = re.match('[0-9] h',m) # 1 digit
            ind = m.find('h')
            if h2 != None:
                y = int(m[ind-3] + m[ind-2])
                time += y*60
            elif h1 != None:
                y = int(m[ind-2])
                time += y*60
        if 'min' in m:
            minutes3 = re.search('[0-9][0-9][0-9] min',m) # 3 digit
            minutes2 = re.search('[0-9][0-9] min',m) # 2 digit
            minutes1 = re.search('[0-9] min',m) # 1 digit
            ind = m.find('min')
            if minutes3 != None:
                y = int(m[ind-4]+m[ind-3]+m[ind-2])
                time += y
            elif minutes2 != None:
                y = int(m[ind-3]+m[ind-2])
                time += y
            elif minutes1 != None:
                y = int(m[ind-2])
                time += y
        movies['RuntimeInMinutes'][i] = time
```

```python
movies['imdbVotesNormalized'] = x
```

```python
for i in range(n):
    m = movies['imdbVotes'][i]
    if type(m) != float:
        if ',' in m:
            l = m
            y = ''
            for j in l:
                if j != ',':
                    y+=j
            movies['imdbVotesNormalized'][i] = int(y)
        else:
            movies['imdbVotesNormalized'][i] = int(m)
```
